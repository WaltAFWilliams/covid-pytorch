```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import seaborn as sns
import cv2 as cv
```

```python
labels = ['normal', 'lung_opacity', 'pneumonia', 'covid']
df = pd.read_csv('X_Ray_Data.csv')
df.head()
```

# **Exploratory Data Analysis**

```python
fig, ax = plt.subplots()
ax.set_xticks(range(4))
ax.set_xticklabels(labels)
ax.set_title(label='Number of Images per Class')
p1 = ax.bar(x=range(4), height=np.array(df.label.value_counts()))
ax.bar_label(p1, label_type='edge')
```

### **Visualize Images**

```python
def visualize_images(kind):
    # read image
    # Create matplotlib figure
    # plot several images in the figure
    
    label = labels.index(kind)
    paths = df[df['label'] == label].sample(frac=1)['path'].values
    plt.figure(figsize=(20,12))
    
    for n in range(16):
        img = cv.cvtColor(cv.imread(paths[n]), cv.COLOR_BGR2RGB)
        plt.subplot(4,4,n+1)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(img)
        
    plt.tight_layout()
```

## No Disease

```python
visualize_images('normal')
```

## Covid

```python
visualize_images('covid')
```

## Lung Opacity

```python
visualize_images('lung_opacity')
```

## Pneumonia

```python
visualize_images('pneumonia')
```
